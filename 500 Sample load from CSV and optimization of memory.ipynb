# Function that update the table types based on max values from columns, Needs tunning copy from stackoverflow with updates

```python
from pandas.api.types import is_numeric_dtype
def Change_dtypes(df):
    has_decimal = 0
    for col in df.columns:
        # check if is numercio
        if (is_numeric_dtype(df[col])):
            #check limits
            col_min = df[col].min()
            col_max = df[col].max()
            
            bytes = 64
            if ((col_min > -2147483648) & (col_max < 2147483648)):
                bytes = 32
            if ((col_min > -32768) & (col_max < 32768)):
                bytes = 16        
            if ((col_min > -128) & (col_max < 128)):
                bytes = 8
            if ( any(df[col]%1!=0) ):
                has_decimal == 1
                if (bytes == 8):
                    bytes = 16
                type = 'float'+str(bytes)
                print(type)
                #update type col
                df[col] = pd.to_numeric(df[col], downcast='float')
            else:
                type = 'int'+str(bytes)
                print(col, ' type ' , df[col].dtype )
                #update type col
                df[col] = pd.to_numeric(df[col], downcast='integer')
```
